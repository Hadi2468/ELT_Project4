# Project 4: Validation of S3 Bronze Layer Data

```python
import pandas as pd
import json
import glob
```

## media

```python
# Read JSON files
def read_file(path):
    df = (
        pd.read_json(path, typ="series")
          .rename("value")
          .to_frame()
    )
    return df

for file in glob.glob("../data/Bronze/media/*.json"):
    print(file)
    df = read_file(file)
    display(df)
```

## media_engagement

```python
# Read JSON files
def read_json_file(path):
    with open(path) as f:
        data = json.load(f)
    df = pd.DataFrame({
        "engagement_data": data["engagement_data"],
        "rewatch_data": data["rewatch_data"]
    })
    df["engagement"] = data["engagement"]
    return df

for file in glob.glob("../data/Bronze/media_engagement/*.json"):
    print(file)
    df = read_json_file(file)
    display(df)
    df.info()
```
